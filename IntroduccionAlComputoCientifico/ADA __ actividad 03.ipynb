<p  align="justify" style="margin-left: 10%; margin-right: 10%; font-family: 'Cascadia Mono'; font-size: 16px; line-height: 2; font-style: italic;"><span style="color: orange;">Introducción. </span>En el ámbito de las aproximaciones numéricas, la medición y comprensión de errores sirven para validar métodos computacionales. Además, el error absoluto junto con el error relativo son indicadores esenciales para cuantificar la discrepancia entre una aproximación con el valor real de una cantidad. <span style="color: orange;">El error absoluto (Ep)</span> cuantifica la magnitud de la diferencia entre la aproximación y el valor real, mientras que <span style="color: orange;">el error relativo (Rp)</span> proporciona una medida relativa de precisión al considerar el error absoluto en relación con el valor absoluto del resultado original. Esto se centra en la aproximación de la integral definida <span style="color: orange;">∫(0 to 1/2) e^(x^2)</span> mediante la serie de Taylor truncada <span style="color: orange;">Ps(x) = 1 + x^2 + (x^4)/2! + (x^6)/3! + (x^8)/4!</span>. Además, se introduce el concepto de error computacional al examinar la función <span style="color: orange;">f(x) = e^(5x)</span> y utilizar diferencias finitas para aproximar su derivada en un punto específico. La representación gráfica del error computacional en función del tamaño del <span style="color: orange;">paso (h)</span> revela información valiosa sobre la precisión del método utilizado.</p>

<p  align="justify" style="margin-left: 10%; margin-right: 10%; font-family: 'Cascadia Mono'; font-size: 16px; line-height: 2;"><span style="color: orange;">El error absoluto de una aproximación </span>se refiere a la magnitud de la diferencia entre el valor aproximado de una cantidad junto con su valor real o verdadero. Es una medida de cuánto se desvía la aproximación del valor real</p>

$$ E_p=\left|\:p-p°\:\right|=\left|\:Aproximación-Valor\:Real\:\right| $$

<p  align="justify" style="margin-left: 10%; margin-right: 10%; font-family: 'Cascadia Mono'; font-size: 16px; line-height: 2;"><span style="color: orange;">El error relativo </span>es otra medida que se utiliza para expresar la precisión de una aproximación en relación con el valor real de una cantidad</p>

$$ R_p=\frac{\left|\:p-p°\:\right|}{\left|\:p\:\right|}=\frac{\left|\:Aproximación-Valor\:Real\:\right|}{\left|\:Valor\:Real\:\right|} $$

<p  align="justify" style="margin-left: 10%; margin-right: 10%; font-family: 'Cascadia Mono'; font-size: 16px; line-height: 2;">Sabiendo que</p>

$$ \int _0^{\frac{1}{2}}e^{x^2}=0.544987104184 $$

$$ e^{x^2} $$

$$ P_s\left(\:x\:\right)=1+x^2+\frac{x^4}{2!}+\frac{x^6}{3!}+\frac{x^8}{4!} $$

$$ \int _0^{\frac{1}{2}}1+x^2+\frac{x^4}{2!}+\frac{x^6}{3!}+\frac{x^8}{4!}dx=\frac{2109491}{3870720} $$

<p  align="justify" style="margin-left: 10%; margin-right: 10%; font-family: 'Cascadia Mono'; font-size: 16px; line-height: 2;"><span style="color: orange;">El error absoluto</span> se obtiene calculando la diferencia entre ambos valores y tomando su <span style="color: orange;">valor absoluto</span>, mientras que <span style="color: orange;">el error relativo</span> se obtiene dividiendo <span style="color: orange;">el error absoluto</span> entre el valor absoluto del resultado original.</p>

$$ E_p=\left|\:0.544987104184-\frac{2109491}{3870720}\:\right|=\frac{16259}{2419200000} $$

<p  align="justify" style="margin-left: 10%; margin-right: 10%; font-family: 'Cascadia Mono'; font-size: 16px; line-height: 2;"><span style="color: orange;">Un error absoluto</span> más bajo sugiere una mejor aproximación, y <span style="color: orange;">el error relativo</span> proporciona una medida relativa de la precisión en comparación con el tamaño del valor original.</p>

$$ R_p=\frac{\left|\:0.544987104184-\frac{2109491}{3870720}\:\right|}{\left|\:0.544987104184\:\right|}=\frac{16259}{1318415616} $$

<p  align="justify" style="margin-left: 10%; margin-right: 10%; font-family: 'Cascadia Mono'; font-size: 16px; line-height: 2;">Encuentre la mínima cota de error.</p>

<p  align="justify" style="margin-left: 10%; margin-right: 10%; font-family: 'Cascadia Mono'; font-size: 16px; line-height: 2; font-style: italic;"><span style="color: orange;">Aproximación de Diferencias Finitas.</span> El código analiza la aproximación de la derivada numérica de la función exponencial <span style="color: orange;">\(e^{5x}\) en \(x = 1\)</span> utilizando diferencias finitas. Se exploran diferentes tamaños de paso <span style="color: orange;">(\(h\))</span> en una escala logarítmica, calculando las derivadas numéricas y evaluando los errores respecto a la derivada exacta. La posición del mínimo error se identifica y se destaca en el gráfico <span style="color: orange;">log-log</span>. Además, proporciona una visualización que muestra cómo el error computacional varía con diferentes tamaños de paso, destacando la mínima cota de error y su correspondiente tamaño de paso. Este enfoque permite analizar la precisión de la aproximación numérica en función del tamaño del paso en el contexto de la función exponencial dada.</p>

```python
from math import *
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return exp(5*x)

def fx(x):
    return 5 * exp(5*x)

h_vector = [10**(-i) for i in np.arange(0, 16, 0.9)]
nh = len(h_vector)
FD = np.zeros(nh)
exact = np.zeros(nh)
FD1 = np.zeros(nh)

x = 1

df_e = fx(x)

for i in range(nh):
    h = 10**(-i)
    h_vector[i] = h
    
    df1 = (f(x+h) - f(x)) / h
    FD[i] = df1
    exact[i] = df_e
    FD1[i] = abs(FD[i] - exact[i])

min_error_index = np.argmin(FD1)
min_error_h = h_vector[min_error_index]
min_error_value = FD1[min_error_index]

plt.figure()
plt.loglog(h_vector, FD1, label='Error')
plt.scatter(min_error_h, min_error_value, color='red', label='Mínima Cota Error', zorder=5)
plt.text(min_error_h, min_error_value, f'Mínima Cota Error\nh = {min_error_h:.2e}', color='red', fontsize=10, va='bottom', ha='right')
plt.xlim(10**-16, 10**0)
plt.ylim(10**-18, 10**2)
plt.legend()
plt.grid()
plt.show()
```

<p  align="justify" style="margin-left: 10%; margin-right: 10%; font-family: 'Cascadia Mono'; font-size: 16px; line-height: 2; font-style: italic;"><span style="color: orange;">Diferencias Finitas.</span> El código evalúa y compara dos fórmulas de aproximación de derivadas numéricas <span style="color: orange;">(denominadas Formula 1 y Formula 2)</span> en relación con la derivada exacta de la función tangente en <span style="color: orange;">x = 1</span>. Utiliza el método de ajuste de curvas para determinar la tasa de convergencia de errores en función del tamaño del paso (h). La función <span style="color: orange;">compute_fd</span> calcula las derivadas numéricas de primer y tercer orden con respecto a h. Luego, se calculan los errores relativos entre las derivadas numéricas y la derivada exacta para diferentes valores de h. Utilizando regresión polinómica, se ajustan curvas a los datos logarítmicos de los errores. Los coeficientes de las curvas ajustadas proporcionan información sobre la tasa de convergencia. Además, se imprimen y visualizan los resultados mediante un gráfico <span style="color: orange;">log-log</span> que muestra las fórmulas, las curvas de ajuste y sus respectivos exponentes. Este enfoque permite analizar la precisión y la convergencia de las aproximaciones numéricas en función del tamaño del paso.</p>

```python
from mpmath import *
from math import *
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return tan(x)

def fx(x):
    return sec(x) * sec(x)

def compute_fd(h, x):
    df1 = (f(x + h) - f(x - h)) / (2 * h)
    df2 = (-11 * f(x) + 18 * f(x + h) - 9 * f(x + 2 * h) + 2 * f(x + 3 * h)) / (6 * h)
    return df1, df2

h_vector = np.logspace(-16, 0, num=34)
exact_solution = fx(1)
FD1 = np.zeros_like(h_vector)
FD3 = np.zeros_like(h_vector)

for i, h in enumerate(h_vector):
    df1, df2 = compute_fd(h, 1)
    FD1[i] = abs((df1 - exact_solution) / exact_solution)
    FD3[i] = abs((df2 - exact_solution) / exact_solution)

coeff_fd1 = np.polyfit(np.log(h_vector[:11]), np.log(FD1[:11]), 1)
coeff_fd3 = np.polyfit(np.log(h_vector[:7]), np.log(FD3[:7]), 1)

error_fit_cen = exp(coeff_fd1[1]) * (h_vector[:11] ** coeff_fd1[0])
error_fit_for = exp(coeff_fd3[1]) * (h_vector[:7] ** coeff_fd3[0])

plt.figure()
plt.loglog(h_vector, FD1, '--r', label='Formula 1')
plt.loglog(h_vector, FD3, '--b', label='Formula 2')
plt.loglog(h_vector[:11], error_fit_cen, '--g', label='Curve Fit_c')
plt.loglog(h_vector[:7], error_fit_for, '--k', label='Curve Fit_f')
plt.xlim(10**-16, 1)
plt.ylim(10**-12, 100)
plt.legend()
plt.xlabel('Step h')
plt.ylabel('Error')
plt.grid()
plt.show()
```

<p  align="justify" style="margin-left: 10%; margin-right: 10%; font-family: 'Cascadia Mono'; font-size: 16px; line-height: 2; font-style: italic;"><span style="color: orange;">Arquímedes. </span>El código en Python define una función llamada <span style="color: orange;">calcular_areas_y_c</span>, que realiza iteraciones para calcular y almacenar en listas los valores de a, b y c según ciertas fórmulas geométricas. Luego, con un número específico de iteraciones <span style="color: orange;">(n_valor establecido en 40)</span>, se imprime de manera ordenada en columnas los valores de a, b y c para cada iteración. Cada iteración actualiza los valores basándose en los cálculos geométricos previos. Este enfoque proporciona una visión detallada de cómo evolucionan los valores de las variables a medida que progresa el proceso iterativo.</p>

```python
import math

def calcular_areas_y_c(n):
    a = [0] * (n + 1)
    b = [0] * (n + 1)
    c = [0] * (n + 1)

    a[0] = (3 / 4) * math.sqrt(3)
    b[0] = 3 * math.sqrt(3)
    c[0] = (a[0] + b[0]) / 2

    for i in range(0, n):
        a[i + 1] = math.sqrt(a[i] * b[i])
        b[i + 1] = 2 / ((1 / a[i + 1]) + (1 / b[i]))
        c[i] = (a[i] + b[i]) / 2
    return a, b, c

n_valor = 40
a_n, b_n, c_n = calcular_areas_y_c(n_valor)

print("Iteración\t a\t\t b\t\t c")
for i in range(n_valor + 1):
    print(f"{i}\t\t {a_n[i]:.6f}\t {b_n[i]:.6f}\t {c_n[i]:.6f}")
```

<p  align="justify" style="margin-left: 10%; margin-right: 10%; font-family: 'Cascadia Mono'; font-size: 16px; line-height: 2; font-style: italic;"><span style="color: orange;">Arquímedes. </span>El código en Python define una función llamada <span style="color: orange;">calcular_areas_y_c</span>, que realiza iteraciones para calcular áreas y longitudes de segmentos geométricos. Luego, se calculan errores respecto a π para dos secuencias resultantes, <span style="color: orange;">a_n</span> y <span style="color: orange;">c_n</span>, con 40 iteraciones. Estos errores se representan en una gráfica logarítmica utilizando la biblioteca <span style="color: orange;">matplotlib</span>. La visualización permite observar cómo los errores convergen hacia cero con el aumento de iteraciones, evaluando así la precisión de los cálculos en relación con el valor de <span style="color: orange;">π</span>.</p>

```python
import math
import matplotlib.pyplot as plt

def calcular_areas_y_c(n):
    a = [0] * (n + 1)
    b = [0] * (n + 1)
    c = [0] * (n + 1)

    a[0] = (3 / 4) * math.sqrt(3)
    b[0] = 3 * math.sqrt(3)
    c[0] = (a[0] + b[0]) / 2

    for i in range(0, n):
        a[i + 1] = math.sqrt(a[i] * b[i])
        b[i + 1] = 2 / ((1 / a[i + 1]) + (1 / b[i]))
        c[i] = (a[i] + b[i]) / 2
    return a, c

n_valor = 40
a_n, c_n = calcular_areas_y_c(n_valor)

error_a_n = [abs(a_i - math.pi) for a_i in a_n]
error_c_n = [abs(c_i - math.pi) for c_i in c_n]

plt.semilogy(error_a_n, label='|a_n - π|', linestyle='-', marker='o')
plt.semilogy(error_c_n, label='|c_n - π|', linestyle='-', marker='x')
plt.legend()
plt.show()
```
