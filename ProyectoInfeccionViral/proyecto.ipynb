**Introducción**

Los mecanismos activos del sistema inmunológico contra los virus se pueden clasificar en dos tipos principales: aquellos que implican la destrucción directa de las células infectadas y aquellos que actúan de otras maneras. Hemos empleado modelos matemáticos para explorar qué tipo de respuesta inmunitaria es necesaria para combatir diferentes tipos de infecciones virales. En nuestro estudio, encontramos que la importancia relativa de estos dos tipos de mecanismos depende de la capacidad del virus para causar daño celular en comparación con su velocidad de replicación. Cuando el daño celular es bajo en relación con la velocidad de replicación del virus, nuestro modelo sugiere que una combinación de mecanismos de destrucción celular y otros mecanismos inmunitarios puede ser necesaria para resolver la infección, especialmente si el virus se replica rápidamente. Sin embargo, si el daño celular causado por el virus es alto en comparación con su velocidad de replicación, entonces los mecanismos de destrucción celular y otros pueden ser capaces de resolver la infección de manera independiente. Hemos aplicado estos hallazgos al análisis de infecciones virales específicas y hemos utilizado nuestro modelo para interpretar datos empíricos.

El modelo se expresa mediante las siguientes ecuaciones:
$$
\begin{align*}
    & \frac{x\left(t\right)}{dt}=s-d\cdot x\left(t\right)-\frac{\beta \cdot x\left(t\right)\cdot y\left(t\right)}{1+q\cdot z\left(t\right)}\\
    & \frac{y\left(t\right)}{dt}=\frac{\beta \cdot x\left(t\right)\cdot y\left(t\right)}{1+q\cdot z\left(t\right)}-a\cdot y\left(t\right)-p\cdot y\left(t\right)\cdot z\left(t\right)\\
    & \frac{z\left(t\right)}{dt}=c\cdot y\left(t\right)-b\cdot z\left(t\right)\\
\end{align*}
$$

O descrito de otra forma : 
$$
\begin{align*}
    & x'=\lambda -dx-\frac{\beta \cdot x\cdot y}{1+q\cdot z} \\
    & y'=\frac{\beta \cdot x\cdot y}{1+q\cdot z}-a\cdot y-p\cdot y\cdot z \\
    & z'=cy-bz  \\
\end{align*}
$$

En estas ecuaciones, los términos $ \frac{x(t)}{dt} $ , $ \frac{y(t)}{dt} $ , y $ \frac{z(t)}{dt} $ representan las tasas de cambio instantáneas de las poblaciones de células hospedadoras susceptibles, virus y respuestas inmunes, respectivamente. Aquí están los parámetros asociados con cada ecuación:

1. Para la ecuación de $ x(t) $:
   - $ s $: Tasa de generación de células hospedadoras susceptibles.
   - $ d $: Tasa de muerte de las células hospedadoras susceptibles.
   - $ \beta $: Tasa de infección viral. Es la tasa a la cual las células hospedadoras susceptibles se infectan con el virus.
   - $ q $: Eficacia del componente no lírico del sistema inmune. Cuanto mayor sea $ q $, más efectivo es el sistema inmune para inhibir la replicación viral.
   
2. Para la ecuación de $ y(t) $:
   - $ \beta $: Tasa de infección viral.
   - $ a $: Tasa de muerte de las células infectadas.
   - $ p $: Tasa de eliminación de las células infectadas por el sistema inmune. Este parámetro modela la actividad efectora lírica del sistema inmune.

3. Para la ecuación de $ z(t) $:
   - $ c $: Tasa de crecimiento de las respuestas inmunes. Esta tasa se incrementa proporcionalmente al número de células infectadas.
   - $ b $: Tasa de decaimiento de las respuestas inmunes. Esta tasa modela la disminución en la fuerza de las respuestas inmunes con el tiempo.

Estos parámetros controlan cómo cambian las poblaciones de células hospedadoras susceptibles, virus y respuestas inmunes en función del tiempo, según los procesos biológicos y la interacción entre ellos.

Se encontró que el modelo exhibe diferentes tipos de comportamiento, incluyendo puntos de equilibrio estables e inestables, así como ciclos límite. Estos hallazgos sugieren que la interacción entre la infección viral y la respuesta inmune puede dar lugar a una variedad de resultados, desde la erradicación de la infección hasta la persistencia crónica.

En resumen, este estudio proporciona una comprensión más profunda de la dinámica de la infección viral y la respuesta inmune, y destaca la importancia de considerar múltiples factores en el diseño de estrategias de control y prevención de enfermedades.
